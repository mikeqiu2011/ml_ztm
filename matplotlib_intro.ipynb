## matplotlib example workflow

```python
# 0. import
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup a plot
fit, ax = plt.subplots(figsize=(3,3))

# 3. plot the data
ax.plot(x, y)

# 4. customize plot
ax.set(title='simple plot',
      xlabel='x-axis',
      ylabel='y-axis',
      )

# 5. save and show
fig.savefig('img/sample_plt.png')
```

### two options for subplot

```python
# subplot option 1
fit, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# plot to each ax
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(prices.keys(), prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# plot to each ax
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(prices.keys(), prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### which one shall you use, pyplot or OO
* when simple and quick, use pyplot
* if simple does not work well, for advanced, use OO

```python
df.head(2)
```

```python
over_50 = df.loc[df.age>50]
len(over_50)
```

```python
# pyplot method which not work well in below use case
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
# use OO
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)

ax.set_xlim(45, 100)
```

```python
# subplot of age, chol, thalach
fig, ax = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10,6))

scatter = ax[0].scatter(x=over_50['age'],
               y=over_50['chol'],
               c=over_50['target'])

ax[0].set(title='Heart disease and cholesterol level',
         xlabel='Age',
         ylabel='cholesterol',
         )

ax[0].legend(*scatter.legend_elements(), title='Target')
```
